```python
cancer=datasets.load_breast_cancer()
```

```python
cancer_data=cancer.data
```

```python
cancer_data = pd.DataFrame(cancer_data, columns = cancer.feature_names)
cancer_data['class'] = cancer.target
```

```python
#heat map to find correlation
fig, ax = plt.subplots(figsize=(12,10))
sns.heatmap(cancer_data.corr(),ax=ax)
```

```python
#to find outliers
for column in cancer_data:
        plt.figure(figsize=(5,2))
        sns.boxplot(x=column,data=cancer_data,palette="colorblind")
```

```python
#removing the outliers
Q1=cancer_data.quantile(0.25)
Q3=cancer_data.quantile(0.75)
IQR=Q3-Q1
```

```python
cdata_out = cancer_data[~((cancer_data < (Q1 - (1.5 * IQR))) |(cancer_data > (Q3 + (1.5 * IQR)))).any(axis=1)]
cancer_data.shape,cdata_out.shape
```

```python
cdata_out.head()
```

```python
# Define X and y

X = cdata_out.drop(['class'], axis=1)
y = cdata_out['class'] 
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# train model using SVM algorithm
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score,accuracy_score,make_scorer
accuracy=[]
svc_model = SVC(gamma='auto')
svc_model.fit(X_train, y_train)
```

```python
# predicting with the trained model
y_predict = svc_model.predict(X_test)
#creating a confusion matrix
cm = confusion_matrix(y_test, y_predict)
# plotting the heatmap
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, y_predict))

```

```python
accuracy_score(y_test,y_predict)
```

```python
temp=accuracy_score(y_test,y_predict)
accuracy.append(temp)
```

```python
#KNN

#from sklearn.model_selection import cross_validate
from sklearn.metrics import roc_auc_score,accuracy_score,make_scorer
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=9,n_jobs=-1)
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
```

```python
accuracy_score(y_test, predict)

```
